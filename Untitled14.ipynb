```python
import numpy as np
import pandas as pd
tech = pd.read_csv("tech.csv",index_col=0,parse_dates=True);
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
tech
```

```python
apple= tech.loc[tech.name=='AAPL','open'].rename('Apple')
facebook= tech.loc[tech.name=='FB','open'].rename('Facebook')
google= tech.loc[tech.name=='GOOGL','open'].rename('Google')
```

```python
dfs= pd.concat([apple,facebook,google],axis=1)
```

```python
dfs
```

```python
dfs.plot();
```

```python
dfs.plot(subplots=True,linewidth=0.7,figsize=(13,5));
```

```python
ax=tech.name.value_counts().plot(kind='bar',figsize=(6,5));
ax.set_xlabel('tech name');
ax.set_ylabel('name comes up');
```
